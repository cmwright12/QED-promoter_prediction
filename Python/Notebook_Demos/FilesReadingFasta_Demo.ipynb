# Reading FASTA files

Reading *fasta* files is an important part of bioinformatics. Let's read a fasta file and output just the beginning of our data.

```python
phix = "files/phix.fa"
sample = "files/fasta_sample.fna"
```

```python
with open(phix) as fasta_file:
    data = fasta_file.read()
    print(data[:500])
    print("The output is of type {}. It is of length {}.".format(type(data), len(data)))
```

We could've specified we only wanted the first 500 characters in our file using `read`.

```python
with open("files/phix.fa") as f:
    data = f.read(500)
    print(data)
```

Pretty simple, right? But what if I only want the sequence? That means I need to ignore the first row. Furthermore, the rest looks like a continuous string, but they are all actually separate lines because of the newline character.

```python
with open(phix) as f:
    # one way to ignore the first line
    header = f.readline().rstrip()
    data = f.read()
    # does it contain any newline characters?
    print('\n' in data)
```

We need to loop through the data, strip off the trailing newline, and append each line to form one continuous string.

```python
with open(phix) as f:
    header = f.readline().rstrip()
    data = f.read()
    seq = ""
    for line in data:
        line = line.rstrip()
        seq = seq + line
    print(seq)
    # any newline characters?
    print('\n' in seq)
```

An alternative way is to just read in all of the data as a list. If we want to ignore the first row, we just start our loop from the second row. But the previous way is better if you need to capture any sequence information. In this example, we don't want or need it.

```python
with open(sample) as f:
    data = f.readlines()
    seq = ""
    for line in data[1:]:
        line = line.rstrip()
        seq = seq + line
    print(seq)
    # any newline characters?
    print('\n' in seq)
```

Function that takes in fasta filename and outputs only the sequence:

```python
def get_seq_from_fasta(filename):
    with open(filename) as f:
        data = f.readlines()
        seq = ""
        for line in data[1:]:
            line = line.rstrip()
            seq = seq + line
    return seq
```

```python
myseq = get_seq_from_fasta(phix)
myseq
```

** Option: If you know about list comprehensions you can do the following.

```python
def get_seq_from_fasta2(filename):
    with open(filename) as f:
        data = f.readlines()
        seq = [line.rstrip() for line in data[1:]]
        seq = "".join(seq)
    return seq
```

```python
myseq = get_seq_from_fasta2(phix)
myseq
```

```python
%timeit get_seq_from_fasta(phix)
```

```python
%timeit get_seq_from_fasta2(phix)
```

In this case, not a big time difference. It would be interesting to compare the times for much larger files.
